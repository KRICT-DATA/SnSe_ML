```python
import numpy as np
import copy

import matplotlib.pyplot as plt
import xgboost as xgb
import pandas as pd
from joblib import dump, load


from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score


import util
from util import KFold_seed
from util import cut_feature
from util import calculate_ZT
from util import dopant_list
from util import scatterplot
from util import temp_plot

%matplotlib inline



measureset=['electrical_conductivity','seebeck_coeff','thermal_conductivity']
#given feature set
column=['direction','temp',\
          'vbm_e','vbm_e2','vbm_e3',\
          'cbm_e','cbm_e2','cbm_e3','magnetism','deep_level',\
          'vbm_org_1', 'vbm_org_2','vbm_org_3','cbm_org_1','cbm_org_2','cbm_org_3',\
          'vbm_m','vbm_m2','vbm_m3','cbm_m','cbm_m2','cbm_m3',\
          'vbm_e_sn','vbm_e2_sn','vbm_e3_sn','cbm_e_sn','cbm_e2_sn','cbm_e3_sn',\
          'vbm_e_se','vbm_e2_se','vbm_e3_se','cbm_e_se','cbm_e2_se','cbm_e3_se',\
          'vbm_m_sn', 'vbm_m2_sn','vbm_m3_sn','cbm_m_sn','cbm_m2_sn','cbm_m3_sn',\
          'vbm_m_se', 'vbm_m2_se','vbm_m3_se','cbm_m_se','cbm_m2_se','cbm_m3_se',\
        'bandgap','bandgap_sn', 'bandgap_se', 'bandgap_rate',\
          'ion','covalent','vdw','atomic_radius','pauling_en','en_allen','polarizability','melting_point',\
          'boiling_point','density','specific_heat','thermal_conductivity',\
        'Sn_ratio','Se_ratio','Na_ratio','Cl_ratio','dopant_num', 'dopant_ratio']  

column2 = ['id'] + column
featureset= [['direction','temp',\
          'Sn_ratio','Se_ratio','Na_ratio','Cl_ratio','dopant_num', 'dopant_ratio'],\
            ['direction','temp',\
          'ion','covalent','vdw','atomic_radius','pauling_en','en_allen','polarizability','melting_point',\
          'boiling_point','density','specific_heat','thermal_conductivity',\
          'Sn_ratio','Se_ratio','Na_ratio','Cl_ratio','dopant_num', 'dopant_ratio'],\
            ['direction','temp',\
             'magnetism','deep_level',\
          'vbm_m','vbm_m2','vbm_m3','cbm_m','cbm_m2','cbm_m3',\
         'vbm_org_1', 'vbm_org_2','vbm_org_3','cbm_org_1','cbm_org_2','cbm_org_3',\
          'vbm_e_sn','vbm_e2_sn','vbm_e3_sn','cbm_e_sn','cbm_e2_sn','cbm_e3_sn',\
          'vbm_e_se','vbm_e2_se','vbm_e3_se','cbm_e_se','cbm_e2_se','cbm_e3_se',\
          'vbm_m_sn', 'vbm_m2_sn','vbm_m3_sn','cbm_m_sn','cbm_m2_sn','cbm_m3_sn',\
          'vbm_m_se', 'vbm_m2_se','vbm_m3_se','cbm_m_se','cbm_m2_se','cbm_m3_se',\
          'bandgap','bandgap_sn', 'bandgap_se', 'bandgap_rate',\
          'Sn_ratio','Se_ratio','Na_ratio','Cl_ratio','dopant_num', 'dopant_ratio'],\
            ['direction','temp',\
              'magnetism','deep_level',\
          'vbm_m','vbm_m2','vbm_m3','cbm_m','cbm_m2','cbm_m3',\
         'vbm_org_1', 'vbm_org_2','vbm_org_3','cbm_org_1','cbm_org_2','cbm_org_3',\
          'vbm_e_sn','vbm_e2_sn','vbm_e3_sn','cbm_e_sn','cbm_e2_sn','cbm_e3_sn',\
          'vbm_e_se','vbm_e2_se','vbm_e3_se','cbm_e_se','cbm_e2_se','cbm_e3_se',\
          'vbm_m_sn', 'vbm_m2_sn','vbm_m3_sn','cbm_m_sn','cbm_m2_sn','cbm_m3_sn',\
          'vbm_m_se', 'vbm_m2_se','vbm_m3_se','cbm_m_se','cbm_m2_se','cbm_m3_se',\
          'bandgap','bandgap_sn', 'bandgap_se', 'bandgap_rate',\
           'ion','covalent','vdw','atomic_radius','pauling_en','en_allen','polarizability','melting_point',\
          'boiling_point','density','specific_heat','thermal_conductivity',\
          'Sn_ratio','Se_ratio','Na_ratio','Cl_ratio','dopant_num', 'dopant_ratio']]

# hyperparameters
num_feats=len(column)
num_folds = 5
id_list=list()
seed=54
iii=0;ggg=0
n_feature=0
loss_data=np.zeros((4,3,4))
zt_list=np.zeros((4,3))

for feature in featureset:
    
    jjj=0
    r2_list=list()
    # load training dataset
    num_folds = 5
    for measure in measureset:
 
        best_loss = 10000000
        load_file = './inputdata_feature/training_features_'+str(measure)+'.csv'
        data_dict = util.load_data(load_file)

        # generate k-fold training dataset
        kfold = KFold_seed(data_dict, num_feats, num_folds, seed,normalization=False)

        filename_best = './outputdata/featuretest/'+str(measure) + '_best_final.joblib'

        id_list.append(kfold.split_dataset())
        
        # define default model
        if measure == 'electrical_conductivity':
            default_xgb = xgb.XGBRegressor(max_depth=5, n_estimators=500)  #XGB
        elif measure == 'seebeck_coeff':
            default_xgb = xgb.XGBRegressor(max_depth=5, n_estimators=400)  #XGB
        elif measure == 'thermal_conductivity':
            default_xgb = xgb.XGBRegressor(max_depth=5, n_estimators=400)  #XGB


        # evaluate model using k-fold cross-validation
        sum_test_mae = 0;    sum_train_mae = 0;    sum_test_r2 = 0;    sum_train_r2 = 0
        total_train_in=[];   total_train_out=[];   total_test_in=[];   total_test_out=[]

        temp = copy.deepcopy(column)
        temp2=[]

        for j in feature:
            temp.remove(j)

        for k in temp:
            temp2.append(column.index(k))
        temp2.sort(reverse=True)

        for k in range(0, num_folds):
            model = copy.deepcopy(default_xgb)
            train_dataset, test_dataset, train_id_dataset, test_id_dataset = kfold.get(k)
            train_data_x = train_dataset[:, :num_feats]
            train_data_y = train_dataset[:, num_feats].reshape(-1, 1)
            test_data_x = test_dataset[:, :num_feats]
            test_data_y = test_dataset[:,num_feats]

            train_data_x_in = copy.deepcopy(train_data_x)
            test_data_x_in = copy.deepcopy(test_data_x)            

            #remove useless feature columns
            for j in temp2:
                train_data_x_in = np.delete(train_data_x_in,j,axis=1)
                test_data_x_in  = np.delete(test_data_x_in,j,axis=1)           

            X_train_model, Y_train_model = shuffle(train_data_x_in,train_data_y, random_state=423)

            model.fit(X_train_model, Y_train_model, eval_metric='mae', eval_set=[(X_train_model,Y_train_model)],verbose=None)  #Xgb

            filename = str(measure) + '_'+str(k)+".joblib"
            filename ='./outputdata/featuretest/'+str(measure) + '_'+str(k)+'_'+'seed_'+str(ggg)+".joblib"

            dump(model, filename)

            
            train_mae, train_r2 = util.pred_train_ML(model,train_data_x_in,train_dataset,total_train_in, total_train_out, num_feats)
            test_mae, test_r2 =   util.pred_test_ML(model,num_feats, test_data_x_in,test_data_y, total_test_in, total_test_out)
            
            sum_test_mae += test_mae
            sum_train_mae += train_mae
            sum_train_r2 += train_r2
            sum_test_r2 += test_r2
            #print('Fold [{}/{}]\tTrain MAE:{:.4F}, Train R2:{:.4F},\tTest MAE: {:.4F}, Test R2: {:.4F}'.format(k + 1, \
            #    num_folds, train_mae, train_r2, test_mae, test_r2))
            if test_mae < best_loss:
                best_loss=test_mae
                dump(model, filename_best)

        # average test in k-fold cross-validation
       # print('Average train loss: {:.4f}, r2: {:.4f}, Average test loss: {:.4f}, r2: {:.4f}'.format(sum_train_mae/num_folds, \
       #         sum_train_r2/num_folds, sum_test_mae / num_folds, sum_test_r2/num_folds))
        print(measure,'{:.4f}, {:.4f}, {:.4f}, {:.4f}'.format(sum_train_mae/num_folds, \
                sum_train_r2/num_folds, sum_test_mae / num_folds, sum_test_r2/num_folds))
        loss_data[n_feature,jjj,0]=sum_train_mae/num_folds
        loss_data[n_feature,jjj,1]=sum_train_r2/num_folds
        loss_data[n_feature,jjj,2]=sum_test_mae/num_folds
        loss_data[n_feature,jjj,3]=sum_test_r2/num_folds
        
        output=pd.DataFrame(total_train_in)
        output.to_csv('./outputdata/featuretest/drawing_train_in_'+str(jjj)+'.csv')
        
        
        output=pd.DataFrame(total_train_out)
        output.to_csv('./outputdata/featuretest/drawing_train_out_'+str(jjj)+'.csv')        
        
        
        output=pd.DataFrame(total_test_in)
        output.to_csv('./outputdata/featuretest/drawing_test_in_'+str(jjj)+'.csv')
        
        
        output=pd.DataFrame(total_test_out)
        output.to_csv('./outputdata/featuretest/drawing_test_out_'+str(jjj)+'.csv')
        r2_list.append([sum_train_r2/num_folds,sum_test_r2/num_folds])
        jjj+=1        
        



    #draw figure
    fig = plt.figure(figsize=(10, 10))
    fig.subplots_adjust(left=0.02,top=0.98,bottom=0.02,right=0.98,wspace=0.01,hspace=0.2)

    jjj=0
    for measure in measureset:
        ax0 = fig.add_subplot(3, 2, 2*jjj+1)
        ax1 = fig.add_subplot(3, 2, 2*jjj+2)        
        total_train_in = pd.read_csv('./outputdata/featuretest/drawing_train_in_'+str(jjj)+'.csv', index_col=0)
        total_train_out = pd.read_csv('./outputdata/featuretest/drawing_train_out_'+str(jjj)+'.csv', index_col=0)
        total_test_in = pd.read_csv('./outputdata/featuretest/drawing_test_in_'+str(jjj)+'.csv', index_col=0)
        total_test_out = pd.read_csv('./outputdata/featuretest/drawing_test_out_'+str(jjj)+'.csv', index_col=0)
        
        util.figsetting(measure, fig, ax0)
        ax0.scatter(total_train_in, total_train_out, alpha=0.2, edgecolors='black')
        ax0.annotate('$R^2$ = {:.2f}'.format(r2_list[jjj][0]),xy=(0.05,0.85),xycoords='axes fraction',size='small',backgroundcolor='white',\
                    fontsize=16)

        util.figsetting(measure, fig, ax1)
        ax1.scatter(total_test_in,total_test_out, alpha=0.2, edgecolors='black', color='red')
        ax1.annotate('$R^2$ = {:.2f}'.format(r2_list[jjj][1]),xy=(0.05,0.85),xycoords='axes fraction',size='small',backgroundcolor='white',\
                    fontsize=16)

        jjj+=1


    plt.show()
    fig.savefig('./outputdata/featuretest/pred_all.pdf', format='pdf')
    
    
    
    load_file = './inputdata_feature/zt_test_features.csv'
    data_dict_2 = util.load_data(load_file)

    thermo  = pd.read_csv('./inputdata/zt_dataset.csv', index_col=0)
    
    target=thermo.K
    num_folds_zt=1
    kfold = KFold_seed(data_dict_2, num_feats, num_folds_zt, seed,normalization=False)
    id_list_zt=kfold.split_dataset_zt()[0]
    print(id_list_zt,len(id_list_zt))
    sample=len(id_list_zt)
    zt_test_in=list()
    zt_test_out=list()
    predicted=np.zeros((len(measureset),thermo.shape[0]))
    iii=0

    for measure in measureset:

        for id in id_list_zt:

            if id in id_list[iii][0]:

                test_dataset = kfold.get_zt(id)
                test_data_x = test_dataset[:, :num_feats]
                test_data_y = test_dataset[:,num_feats]     

                test_data_x_in = copy.deepcopy(test_data_x)            

                #remove useless feature columns
                for j in temp2:
                    test_data_x_in  = np.delete(test_data_x_in,j,axis=1)             

                
                filename = './outputdata/featuretest/'+str(measure) + "_" +  str(0) +'_'+'seed_'+str(ggg)+ ".joblib"
                model = load(filename)        
                test_mae, test_r2 =   util.pred_test_ML(model,num_feats, test_data_x_in,test_data_y, zt_test_in, zt_test_out)


            elif id in id_list[iii][1]:

                test_dataset = kfold.get_zt(id)
                test_data_x = test_dataset[:, :num_feats]
                test_data_y = test_dataset[:,num_feats]     

                test_data_x_in = copy.deepcopy(test_data_x)            

                #remove useless feature columns
                for j in temp2:
                    test_data_x_in  = np.delete(test_data_x_in,j,axis=1)             

                
                filename = './outputdata/featuretest/'+str(measure) + "_" +  str(1) +'_'+'seed_'+str(ggg)+ ".joblib"

                model = load(filename)        
                test_mae, test_r2 =   util.pred_test_ML(model,num_feats, test_data_x_in,test_data_y, zt_test_in, zt_test_out)


            elif id in id_list[iii][2]:

                test_dataset = kfold.get_zt(id)
                test_data_x = test_dataset[:, :num_feats]
                test_data_y = test_dataset[:,num_feats]     

                test_data_x_in = copy.deepcopy(test_data_x)            

                #remove useless feature columns
                for j in temp2:
                    test_data_x_in  = np.delete(test_data_x_in,j,axis=1)             

                
                filename = './outputdata/featuretest/'+str(measure) + "_" +  str(2) +'_'+'seed_'+str(ggg)+ ".joblib"

                model = load(filename)        
                test_mae, test_r2 =   util.pred_test_ML(model,num_feats, test_data_x_in,test_data_y, zt_test_in, zt_test_out)


            elif id in id_list[iii][3]:

                test_dataset = kfold.get_zt(id)
                test_data_x = test_dataset[:, :num_feats]
                test_data_y = test_dataset[:,num_feats]     

                test_data_x_in = copy.deepcopy(test_data_x)            

                #remove useless feature columns
                for j in temp2:
                    test_data_x_in  = np.delete(test_data_x_in,j,axis=1)             

                
                filename = './outputdata/featuretest/'+str(measure) + "_" +  str(3) +'_'+'seed_'+str(ggg)+ ".joblib"

                model = load(filename)        
                test_mae, test_r2 =   util.pred_test_ML(model,num_feats, test_data_x_in,test_data_y, zt_test_in, zt_test_out)

            elif id in id_list[iii][4]:

                test_dataset = kfold.get_zt(id)
                test_data_x = test_dataset[:, :num_feats]
                test_data_y = test_dataset[:,num_feats]     

                test_data_x_in = copy.deepcopy(test_data_x)            

                #remove useless feature columns
                for j in temp2:
                    test_data_x_in  = np.delete(test_data_x_in,j,axis=1)             

                
                filename = './outputdata/featuretest/'+str(measure) + "_" +  str(4) +'_'+'seed_'+str(ggg)+ ".joblib"

                model = load(filename)        
                test_mae, test_r2 =   util.pred_test_ML(model,num_feats, test_data_x_in,test_data_y, zt_test_in, zt_test_out)
            else:

                print(id,'skip')
            #print(id,len(zt_test_in),len(zt_test_out))

        predicted[iii,:]=np.array(zt_test_out)
        test_mae = np.mean(np.abs(np.array(zt_test_in) - np.array(zt_test_out)))
        test_r2 = r2_score(np.array(zt_test_out), np.array(zt_test_in))

        zt_test_in=list()
        zt_test_out=list()
        iii+=1


    temp=thermo.Temp
    ZT=calculate_ZT(predicted,temp,target,thermo, True, './outputdata/featuretest/zt_test.csv')    
    




    X_train=dopant_list(thermo) # dopant list rearrangement
    target = thermo.K.values
    temp = thermo.Temp.values

    scatterplot(target,ZT,temp,'Temperature','./outputdata/featuretest/zt_temp.pdf')


    rmse_train = np.sqrt(mean_squared_error(target, ZT))
    mae_train = mean_absolute_error(target,ZT)

    r2score_train = r2_score(ZT, target)

    print("%.4f" % rmse_train,"%.4f" % mae_train,"%.4f" % r2score_train)
    print()

    zt_list[n_feature][0] = rmse_train
    zt_list[n_feature][1] = mae_train
    zt_list[n_feature][2] = r2score_train
    
    temp_plot(temp,target,ZT,'./outputdata/featuretest/MAE_ZT.pdf')



   
    n_feature +=1
    
print(loss_data)
```
