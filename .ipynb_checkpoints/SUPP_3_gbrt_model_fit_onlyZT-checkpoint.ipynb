```python
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    test=np.abs(y_true - y_pred) / np.average(y_true) / len(y_true) 
    
    return test.sum()
```

```python
import numpy as np
import copy

import matplotlib.pyplot as plt
import xgboost as xgb
import pandas as pd
from joblib import dump, load


from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score


import util
from util import KFold_seed
from util import cut_feature
from util import calculate_ZT
from util import dopant_list
from util import scatterplot
from util import temp_plot

%matplotlib inline





measureset=['zt']

#given feature set
column=['direction','temp',\
          'vbm_e','vbm_e2','vbm_e3',\
          'cbm_e','cbm_e2','cbm_e3','magnetism','deep_level',\
          'vbm_org_1', 'vbm_org_2','vbm_org_3','cbm_org_1','cbm_org_2','cbm_org_3',\
          'vbm_m','vbm_m2','vbm_m3','cbm_m','cbm_m2','cbm_m3',\
          'vbm_e_sn','vbm_e2_sn','vbm_e3_sn','cbm_e_sn','cbm_e2_sn','cbm_e3_sn',\
          'vbm_e_se','vbm_e2_se','vbm_e3_se','cbm_e_se','cbm_e2_se','cbm_e3_se',\
          'vbm_m_sn', 'vbm_m2_sn','vbm_m3_sn','cbm_m_sn','cbm_m2_sn','cbm_m3_sn',\
          'vbm_m_se', 'vbm_m2_se','vbm_m3_se','cbm_m_se','cbm_m2_se','cbm_m3_se',\
        'bandgap','bandgap_sn', 'bandgap_se', 'bandgap_rate',\
          'ion','covalent','vdw','atomic_radius','pauling_en','en_allen','polarizability','melting_point',\
          'boiling_point','density','specific_heat','thermal_conductivity',\
        'Sn_ratio','Se_ratio','Na_ratio','Cl_ratio','dopant_num', 'dopant_ratio']  

# available remove features from the total feature list
feature=['direction','temp',\
          'magnetism','deep_level',\
          'vbm_org_1', 'vbm_org_2','vbm_org_3','cbm_org_1','cbm_org_2','cbm_org_3',\
         'vbm_m','vbm_m2','vbm_m3','cbm_m','cbm_m2','cbm_m3',\
          'vbm_e_sn','vbm_e2_sn','vbm_e3_sn','cbm_e_sn','cbm_e2_sn','cbm_e3_sn',\
          'vbm_e_se','vbm_e2_se','vbm_e3_se','cbm_e_se','cbm_e2_se','cbm_e3_se',\
          'vbm_m_sn', 'vbm_m2_sn','vbm_m3_sn','cbm_m_sn','cbm_m2_sn','cbm_m3_sn',\
          'vbm_m_se', 'vbm_m2_se','vbm_m3_se','cbm_m_se','cbm_m2_se','cbm_m3_se',\
          'bandgap','bandgap_sn', 'bandgap_se', 'bandgap_rate',\
          'ion','covalent','vdw','atomic_radius','pauling_en','en_allen','polarizability','melting_point',\
          'boiling_point','density','specific_heat','thermal_conductivity',\
          'Sn_ratio','Se_ratio','Na_ratio','Cl_ratio','dopant_num', 'dopant_ratio']



# hyperparameters
num_feats=len(column)
num_folds = 5



seedset=[54]


# load training dataset
for ggg, seed in enumerate(seedset):
    measure_num=0
    id_list=list()
    
    for measure in measureset:
        total_temp=list()
        best_loss = 10000000
        load_file = './inputdata_feature/'+str(measure)+'_test_features.csv'
        data_dict = util.load_data(load_file)
        print(measure)
        print('num of features:', len(feature))
        
        # generate k-fold training dataset
        kfold = KFold_seed(data_dict, num_feats, num_folds,seed, False)



        id_list.append(kfold.split_dataset())
 
        print('size of the dataset:', len(id_list[measure_num][0])+len(id_list[measure_num][1])+len(id_list[measure_num][2])+len(id_list[measure_num][3])+len(id_list[measure_num][4]))
 
        measure_num+=1

        # define default model
        if measure == 'electrical_conductivity':
            default_xgb = xgb.XGBRegressor(max_depth=5, n_estimators=500)  
        elif measure == 'seebeck_coeff':
            default_xgb = xgb.XGBRegressor(max_depth=5, n_estimators=400)  
        elif measure == 'thermal_conductivity':
            default_xgb = xgb.XGBRegressor(max_depth=5, n_estimators=400)  
        else:
            default_xgb = xgb.XGBRegressor(max_depth=5, n_estimators=400)  

        # evaluate model using k-fold cross-validation
        sum_test_mae = 0;
        sum_train_mae = 0
        sum_test_r2 = 0
        sum_train_r2 = 0
        total_train_in=[]
        total_train_out=[]
        total_test_in=[]
        total_test_out=[]
        total_test_id=[]
        total_train_id=[]         

        #remove unselected features
        feature_org = copy.deepcopy(column)
        feature_cut =[]

        
        for j in feature:
            feature_org.remove(j)

        for k in feature_org:
            feature_cut.append(column.index(k))
        feature_cut.sort(reverse=True)

        
        
        # construct XGB models
        for k in range(0, num_folds):
            model = copy.deepcopy(default_xgb)
            train_dataset, test_dataset, train_id_dataset, test_id_dataset = kfold.get(k)
            train_data_x = train_dataset[:, :num_feats]
           
            train_data_y = train_dataset[:, num_feats].reshape(-1, 1)
            test_data_x = test_dataset[:, :num_feats]
            test_data_y = test_dataset[:,num_feats]

            train_data_x_in = copy.deepcopy(train_data_x)
            test_data_x_in = copy.deepcopy(test_data_x)            

            #remove useless feature columns
            for j in feature_cut:
                train_data_x_in = np.delete(train_data_x_in,j,axis=1)
                test_data_x_in  = np.delete(test_data_x_in,j,axis=1)           

            #Using Pearson Correlation
            #import seaborn as sns

            #X_train_final=pd.DataFrame(train_data_x_in)
            #X_train_final.columns=feature
            #X_train_final['target']=train_data_y
            #print(X_train_final.shape)

            #plt.figure(figsize=(12,10))
            #cor = X_train_final.corr()
            #sns.heatmap(cor, cmap=plt.cm.Reds)
            #plt.show()

                
                
            #shuffle the data
            X_train_model, Y_train_model = shuffle(train_data_x_in,train_data_y, random_state=423)

            #train the data
            model.fit(X_train_model, Y_train_model, eval_metric='mae', eval_set=[(X_train_model,Y_train_model)],verbose=None)  

            #save the model
            filename ='./outputdata/'+str(measure) + '_'+str(k)+'_'+'seed_'+str(ggg)+".joblib"
            dump(model, filename)

            #error check
            train_mae, train_r2,_ = util.pred_train_ML(model,train_data_x_in,train_dataset,total_train_in, total_train_out, num_feats)
            test_mae, test_r2,_ =   util.pred_test_ML(model,num_feats, test_data_x_in,test_data_y, total_test_in, total_test_out)

            sum_test_mae  += test_mae
            sum_train_mae += train_mae
            sum_train_r2  += train_r2
            sum_test_r2   += test_r2
            print('Fold [{}/{}]\tTrain MAE:{:.4F}, Train R2:{:.4F},\tTest MAE: {:.4F}, Test R2: {:.4F}'.format(k + 1, \
                num_folds, train_mae, train_r2, test_mae, test_r2))

            #save additional information
            total_train_id.extend(train_id_dataset)
            total_test_id.extend(test_id_dataset)              
            total_temp.extend(test_data_x_in[:,1])            
            
            #save the best model for feature importance
            filename_best = './outputdata/'+str(measure) + '_best_final.joblib'
            if test_mae < best_loss:
                best_loss=test_mae
                dump(model, filename_best)

        # average test errors in k-fold cross-validation
        print('Average train loss: {:.4f}, r2: {:.4f}, Average test loss: {:.4f}, r2: {:.4f}'.format(sum_train_mae/num_folds, \
                sum_train_r2/num_folds, sum_test_mae / num_folds, sum_test_r2/num_folds))

        #draw figure
        fig = plt.figure(figsize=(19, 8.5))
        ax0 = fig.add_subplot(2, 1, 1)
        ax1 = fig.add_subplot(2, 2, 2)

        util.figsetting(measure, fig, ax0)
        ax0.scatter(total_train_in, total_train_out, alpha=0.2, edgecolors='black')

        util.figsetting(measure, fig, ax1)
        ax1.scatter(total_test_in,total_test_out, alpha=0.2, edgecolors='black', color='red')
        plt.show()

        fig.savefig('./outputdata/pred_'+str(measure)+'.pdf', format='pdf')

        util.draw_feature_importance(filename_best,feature)
        
        #write predicted values
        targets_all = np.vstack(total_test_in)
        preds_all = np.vstack(total_test_out)
        temp_all = np.vstack(total_temp)
        
        df=pd.DataFrame(data=np.hstack((targets_all, preds_all)), columns=['target', 'predicted'])
        df.loc[:,'Temp']=pd.Series(np.hstack((temp_all)))
        df.loc[:,'abs']=pd.Series(np.hstack((np.abs(targets_all-preds_all))), index=df.index)
        df.loc[:,'multiplication']=pd.Series(np.hstack((targets_all*preds_all)), index=df.index)
        df.loc[:,'id']=pd.Series(total_test_id,  index=df.index)
        df.to_csv('./outputdata/pred_'+str(measure)+'.csv')       
        
        
```
