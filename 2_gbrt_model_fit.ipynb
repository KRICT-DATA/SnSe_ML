```python
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    test=np.sqrt(np.sum((y_true - y_pred)**2)) / np.sqrt(np.sum(y_true**2))
    
    return test.sum()
```
