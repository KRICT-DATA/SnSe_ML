```python
from mendeleev import element
import math
import numpy as np


#

data=np.ndarray((84,12))


    
for i in range(1,84):
    data[i-1,0]=element(i).ionenergies[1]
    data[i-1,1]=element(i).covalent_radius_bragg
    data[i-1,2]=element(i).vdw_radius
    data[i-1,3]=element(i).atomic_radius
    data[i-1,4]=element(i).en_pauling
    data[i-1,5]=element(i).en_allen
    data[i-1,6]=element(i).dipole_polarizability
    data[i-1,7]=element(i).melting_point
    data[i-1,8]=element(i).boiling_point
    data[i-1,9]=element(i).density
    data[i-1,10]=element(i).specific_heat
    data[i-1,11]=element(i).thermal_conductivity
    
    for j in range(12):
        
        if math.isnan(float(data[i-1,j])) ==True:
            data[i-1,j]= 0.0  
            
    
        
    
print(data)
```

```python
import pandas as pd

mendeleev = pd.DataFrame(data=data, columns=['ion','covalent','vdw','atomic_radius','pauling_en','en_allen','polarizability','melting_point',\
          'boiling_point','density','specific_heat','thermal_conductivity'])
```

```python
display(mendeleev)
mendeleev.to_csv('mendeleev.csv')
```
